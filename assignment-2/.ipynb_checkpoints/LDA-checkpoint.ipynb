```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
data = datasets.load_iris()
X, y = data.data, data.target
```

```python
class LDA:
    def __init__(self, nComponents):
        self.nComponents = nComponents
        self.ld = None

    def fit(self, X, y):
        nFeatures = X.shape[1]
        labels = np.unique(y)

        # Within class scatter matrix:
        # SW = sum((X_c - mean_X_c)^2 )

        # Between class scatter:
        # SB = sum( n_c * (mean_X_c - mean_overall)^2 )

        mean_overall = np.mean(X,axis=0)

        # within class scatter matrix        
        SW = np.zeros((nFeatures, nFeatures))
        
        # between class matrix         
        SB = np.zeros((nFeatures, nFeatures))

        
        for c in labels:
            X_c = X[y == c]
            mean_c = np.mean(X_c, axis=0)
            # (4, n_c) * (n_c, 4) = (4,4) -> transpose
            SW += (X_c - mean_c).T.dot((X_c - mean_c))

            # (4, 1) * (1, 4) = (4,4) -> reshape
            n_c = X_c.shape[0]
            mean_diff = (mean_c - mean_overall).reshape(nFeatures, 1)
            SB += n_c * (mean_diff).dot(mean_diff.T)


        # Determine SW^-1 * SB
        A = np.linalg.inv(SW).dot(SB)
        # Get eigenvalues and eigenvectors of SW^-1 * SB
        eValues, eVectors = np.linalg.eig(A)
        # -> eigenvector v = [:,i] column vector, transpose for easier calculations
        # sort eigenvalues high to low
        eVectors = eVectors.T
        idxs = np.argsort(abs(eValues))[::-1]
        eValues = eValues[idxs]
        eVectors = eVectors[idxs]
        # store first n eigenvectors
        self.ld = eVectors[0 : self.nComponents]

    def transform(self, X):
        # project data
        return np.dot(X, self.ld.T)
```

```python
lda = LDA(2)
lda.fit(X, y)
X_projected = lda.transform(X)

print("Shape of X:", X.shape)
print("Shape of transformed X:", X_projected.shape)

x1, x2 = X_projected[:, 0], X_projected[:, 1]

plt.scatter(
    x1, x2, c=y, edgecolor="none", alpha=0.8, cmap=plt.cm.get_cmap("viridis", 3)
)

plt.xlabel("Linear Discriminant 1")
plt.ylabel("Linear Discriminant 2")
plt.colorbar()
plt.show()
```
